**Importing all the necessary libraries** ðŸ“š

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
from matplotlib import cycler
```

**Data Reading and Analysis**

```python
## This is the country list which was defined in Assignment 2
country_codes= [ 'JPN', 'NGA', 'PAK', 'CHN', 'USA', 'IND', 'CAN', 'GHA', 'KEN', 'BGD']

country_groups= { 'JPN':'Asia', 'PAK': 'South Asia', 'IND':'South Asia',
                 'CHN':'Asia', 'USA': 'North America', 'NGA':'Africa', 'CAN': 'North America', 'GHA':'Africa', 'KEN':'Africa', 'BGD': 'South Asia'}
```

```python
#Reading the data and doing analysis to be ready for visulaization in matplotlib
final_new= pd.read_csv('countries_data.csv')
```

```python
final_new= final_new.drop(columns=['Unnamed: 0']) #Deleting the unwanted column
```

```python
df= final_new
```

```python
df_line= df.groupby(['year','country'])['Exports', 'GDP Per Capita (USD)', 'co2 emissions','Access to electricity'].mean().reset_index()
```

```python
df2 = df.set_index('country')
df2
```

```python
#To divide the countries in the regions
region= df2.groupby(country_groups).mean()
```

**Data Visualization in Matplotlib in the form of a dashboard**

```python
## These are taken from the lecture, and the color schemes are changed so to make it according to visual appeal.
colors = cycler('color',
                ['#EE6666', '#3cbd2d', '#47f5ec',
                 '#EECC55', '#d42fad', '#cc74f7',
                '#809185', '#f5a802', '#e502f5',
                '#402633'])
plt.rc('axes', facecolor='white', edgecolor='none',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
## This is the main layout where the GridSpec function is used to define various grids.

fig = plt.figure(figsize=(18, 20))  #This is used to define the figure size of the dashboard.

grid = plt.GridSpec(6, 4, wspace=0.6, hspace=0.9) #A Grid Spec to define a Grid of 5 Rows and 4 columns.

fig.set(facecolor='#f9f5fa')  #This is to set the background color for the dashboard.

## This is to give a title to our dashboard.
fig.suptitle('Dashboard Showing Different Indicators for Countries (2012-2020)', fontsize=20, fontweight='bold', color='#9003a8')


## To add text to the dashboard.
fig.text(0.35,0.95, 'This dashboard is made with Matplotlib GridSpec Function.', fontsize=15)


### Here, different axes are defined, which will then be populated with matplotlib figures ###

ax12 = fig.add_subplot(grid[0,0:])
ax13 = fig.add_subplot(grid[1,0:])

ax21 = fig.add_subplot(grid[2,0])
ax22 = fig.add_subplot(grid[2,1])

ax31 = fig.add_subplot(grid[2,2])
ax32 = fig.add_subplot(grid[2,3])

ax41 = fig.add_subplot(grid[3,0:])

ax51 = fig.add_subplot(grid[4:,0:])


## This is for the first graph, where a for loop is used to make a multiple line chart figure, showing GDP Per Capita of countries.

for i in country_codes:
  a = df[df['country']== i].year
  b = df[df['country']== i]['GDP Per Capita (USD)']
  ax12.plot(a, b, label = i)
 
  ax12.legend(loc='center left', bbox_to_anchor=(1, 0.5))  #The position of legend is set.
  ax12.set_title("GDP Per Capita of Countries since 2012 ") #Title for the figure
  ax12.set_xlabel("Year")  #To set the x-label
  ax12.set_ylabel("GDP Per Capita ") #To set the y-label
  ax12.set(facecolor= '#f9f5fa')  #To set the background color.


##Same is done for the second graph, where the access to electricity for the countries in the list is determined.

for i in country_codes:
  a = df[df['country']== i].year
  b = df[df['country']== i]['Access to electricity']
  ax13.plot(a, b, label = i)

  ax13.legend(loc='center left', bbox_to_anchor=(1, 0.5))
  ax13.set_title("Access to electricity of Countries since 2012 ")
  ax13.set_xlabel("Year")
  ax13.set_ylabel("Access to electricity ")
  ax13.set(facecolor= '#f9f5fa')


## This section is for the pie charts in row 3, where not countries but the overall regions (based on which countries was divided in Assignment 2) were used ###

location = region['country']
data1 = region['Energy use per capita']
ax21.pie(data1, labels = location)
ax21.set_title("Energy User Per Capita by Regions ")
 
data2 = region['co2 emissions']
ax22.pie(data2, labels = location)
ax22.set_title("CO2 Emissions by Regions")


data3 = region['Inflation']
ax31.pie(data3, labels = location)
ax31.set_title("Inflation by Regions")

data4 = region['Access to electricity']
ax32.pie(data4, labels = location)
ax32.set_title("Access to electricity by Regions")


### This is for the scatterplot in Row 4, where a scatterplot for countries and the inflation over the years is made. ###

ax41= sns.scatterplot(ax=ax41, x='year', y='Inflation', hue="country", size="Inflation",
                sizes=(20, 200), data= final_new)
ax41.legend(loc='center left', prop = {'size' : 10}, ncol=2, bbox_to_anchor=(1, 0.5))  #positioning th legend
ax41.set_title("Inflation in countries since 2012")
ax41.set_xlabel("Year")
ax41.set(facecolor= '#f9f5fa')


### A multi-bar chart was made, where the GDP Annula Growth was tracked ####

ax51=sns.barplot(ax= ax51, x= 'year', y='GDP Growth(Annual)', hue='country', data= final_new)
ax51.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax51.set_title("GDP Annual Growth in Countries since 2012")
ax51.set_xlabel("Year")
ax51.annotate('Witness the negative GDP\n Growth in 2020 most \n probably due to COVID', xy =(9, -1),
                xytext =(7, -3), 
                arrowprops = dict(facecolor ='green',
                                  shrink = 0.05),)    ## Used for annotation
ax51.annotate('Only negative growth for Nigeria', xy =(5, -1),
                xytext =(3, -3), 
                arrowprops = dict(facecolor ='red',
                                  shrink = 0.05),)
ax51.set(facecolor= '#f9f5fa')

### Writing some annotations in manual form ####
plt.text(7, 60, 
         r'USA has the highest GDP Per Capita throughout',
         fontdict={'size':9,'color':'r'})    

plt.text(7, 48, 
         r'The rise of Bangladesh to access of electricity is impressive! ',
         fontdict={'size':9,'color':'b'})   


plt.text(7, 22, 
         r'Inflation has been highest in the African region throughout ',
         fontdict={'size':9,'color':'#08000a'}) 


```

**SUMMARY**

The main motive of this dashboard is to display  maximum information for the countries and the indicators provided by the World Bank. A combination of a line chart, bar chart, scatter chart , and pie chart is used to display this information.

This is a dashboard that shows indicators for countries obtained from the World Bank Data. The first figure shows multiple line charts which show the GDP per Capita for countries on the list since 2012. USA has the highest GDP Per Capita throughout while the GDP of some countries is either increasing or decreasing.

The second figure is the Access to Electricity, where advanced countries have full access to electricity. Some countries have an impressive increase, while African countriesâ€™ access to electricity is low. 

In the third row, the pie charts show different indicators for the regions. It shows that Inflammation has been highest in the African Region, Energy Use Per Capita highest in North America, and CO2 emissions highest in North America and South Asia.

The fourth row shows a scatter plot, where countries are shown to have had inflation since 2012. 

The final figure is that of GDP Annual Growth, where most of the countries in the list had negative GDP Growth, including China, Bangladesh , and Ghana.
